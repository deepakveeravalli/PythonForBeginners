## PYTHON LISTS

```python
#Lists are used to store multiple items in a single variable, it is collection of items. 
#Lists are created using square brackets
# Lists are mutable or can be updated

thislist = ["apple", "banana", "cherry"]
print(thislist)
```

```python
#It can store heterogeneous Datatypes:

thislist = ["apple", "banana", "cherry",1,2,True]
print(thislist)
print(type(thislist))
```

```python
#List items are ordered, changeable, and allow duplicate values.List items are indexed, the first item has index [0], the second item has index [1] etc.

#Ordered
#When we say that lists are ordered, it means that the items have a defined order, and that order will not change.

#If you add new items to a list, the new items will be placed at the end of the list.

#Changeable/Mutable
#The list is changeable, meaning that we can change, add, and remove items in a list after it has been created.

#Allow Duplicates

#Since lists are indexed, lists can have items with the same value:

thislist = ["apple", "banana", "cherry", "apple", "cherry"]
print(thislist)
```

```python
#List Length

thislist = ["apple", "banana", "cherry"]
print(len(thislist))
```

```python
#Access Items

thislist = ["apple", "banana", "cherry",1,2,34,True]
print(thislist[1])
```

```python
#Python - Change List Items

thislist = ["apple", "banana", "cherry"]
thislist[1:2] = ["blackcurrant","mango"]
print(thislist)
```

```python
# change the range of items:

thislist[1:3] =['d','e']
print(thislist)
# Change the second value by replacing it with two values:
thislist[1:2] =['d','e']
print(thislist)
```

```python
#Add List Items

#Append

thislist = ["apple", "banana", "cherry"]
print(id(thislist))
thislist.append("orange")
print(thislist)
print(id(thislist))
```

```python
#Extend
thislist = ["apple", "banana", "cherry"]
thislist.append(1)
thislist.append("Kiwi")
print(thislist)
thislist.extend("orange")
print(thislist)
thislist.extend("1")
```

```python
# INSERT
#Add item in the middle of the list

thislist.insert(2,'c')
print(thislist)
```

```python
# Remove Specified Item

#1)Remove Function:
thislist = ["apple", "banana", "cherry", 1 ,2 ]
thislist.remove("banana")
print(thislist)

#2)POP function:
thislist.pop() # removes last item
print(thislist)
thislist.pop(1) # removes item on index 1
print(thislist)
```

# Python Tuples()
- Tuple items are ordered, unchangeable/immutable(Read Only DataType), and allow duplicate values.

- Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created.
- To create a tuple with only one item, you have to add a comma after the item, otherwise Python will not recognize it as a tuple.

<br>Tuples Business scenarios: calendar months
<br>Tuples are more memory efficient than list as list gets extra memory allocation as its mutable. 
<br>Tuples are faster than list



```python
#NOT a tuple
thistuple = ("apple")
print(type(thistuple))
thistuple = tuple(("apple"))
print(type(thistuple))
```

```python
thistuple = ("apple",)
print(type(thistuple))
```

```python
# Tuples can have values of different data types
tup =(1,2,3,"hi",(1,2),[12,4])
print(type(tup))
print(len(tup))
```

```python
# The tuple() Constructor

#It is also possible to use the tuple() constructor to make a tuple.

thistuple = tuple(("apple", "banana", "cherry")) # note the double round-brackets
print(thistuple)
```

```python
# Access Tuple Items: Using Index
print(thistuple[1])

# Once a tuple is created, you cannot change its values. 
```

```python
# But there is a workaround. 

#You can convert the tuple into a list, change the list, and convert the list back into a tuple.

x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)
```

```python
# List inside the tuple are mutable:

tup1 =([1,2],"abc",3.4,(1,2),[1,2],1,2,1)
print(tup1)
tup1[0][0]='a'
print(tup1)
```

```python
#Join Two Tuples

#To join two or more tuples you can use the + operator:

tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2
print(tuple3)
```

```python
participants_list = ['Raju','Deepak','Uday','Shankar','Harish','Kumar']
print(len(participants_list))
```

# 1st way of unpacking a list

LHS should match RHS

```python
par1,par2,par3,par4,par5,par6 = participants_list
print(par1)
print(par6)
```

# 2nd way of unpacking the list

Extract 1st, 2nd or ..n number of elements and assign the remaining to rest operator (*)
LHS variables are assigned values based on the placement of index
the first variable on LHS is assigned first element of the list
similarly, 2nd varibale on LHS will be assigned second element of the list

```python
gold_winner,silver_Winner,bronze_winner,*restofParticipants = participants_list
print(gold_winner)
print(*restofParticipants)
```

# 3rd of list unpacking
Assign first or second or third variable and remaining values in the list except last item

```python

heterogenousList = [1,2,'a','b',4.5,True,'Python']
val1,val2,*middle,last = heterogenousList

print(last)
print(type(last))
print(*middle)
print(type(middle))
print(val1)
print(type(val1))

```

```python
# write a program to find square root of first 10 numbers

squares = []
for n in range(1,11):
    squares.append(n * n)
    
print(squares)
```

```python
squares = [n * n for n in range(1,11)]
print(squares)
```

```python
# write a program to find square root of first 10 even numbers

squares = []
for n in range(1,21):
    if( n % 2 == 0):
        squares.append(n * n)
    
print(squares)
```

```python
squares = [n * n for n in range(1,21) if n % 2 == 0]
print(squares)
```
