```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)
len(thisdict)
```

```python
# Keys cant be duplicated – duplicates are ignored
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2024
}
print(thisdict)
len(thisdict)
```

```python
# values can be duplicated 
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year1": 1964
}
print(thisdict)
```

```python
# The values of item can be of any data types
thisdict = {
  "brand": "Ford",
  "model": [1,2,3],
  "year": {19,64},
  "year1": (1,2,3),
    "dict":{
  "brand": "Ford",
  "model": [1,2,3],
  "year": {19,64},
  "year1": (1,2,3)
}
}
print(thisdict)
```

```python
#dictionary keys can have keys of various data types as long as those data types are immutable.
thisdict = {
  1: "Ford",
  #[1,2,3]: [1,2,3],
  #{19,64}: {19,64},
  (1,2,3): (1,2,3)
}
thisdict1 = {
  1: thisdict,
  #[1,2,3]: [1,2,3],
  #{19,64}: {19,64},
  (1,2,3): (1,2,3)
}
print(thisdict)
print(thisdict1)
```

```python
#Accessing Items/values- Using Index or Get method
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict["model"]
y = thisdict.get("year")
print(x,y)
```

```python
#Get Values
#The values() method will return a list of all the values in the dictionary.
kees = thisdict.keys()
print(kees)
vals = thisdict.values()
print(vals)
```

```python
#Change the value of the Items
thisdict["model"] ='audi'
print(thisdict)
```

```python
# update() method will update the dictionary. 
# update() method will update if the key is already present otherwise will update the dict
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.update({"year": 2020})
thisdict.update({"color": "red"})
print(thisdict)
```

```python
#Adding Items
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["color"] = "red"
print(thisdict)

```

# Dictionary that demoes a person and attributes

Imagine you are developing a form for interview and need all the details of the candidate

1. Name of the dictionary : candidate
2. First Name --> String
3. Last Name --> String
4. Email --> String
5. Phone --> String
6. Gender --> Tuple
7. LinkedIn Profile --> String
8. Qualifications --> Set
9. Skills --> Dictionary 
    9.1 Database Skills --> Set
    9.2 Programming Skills --> Set
    9.3 Managemnet Skills --> Set
    9.4 AI Skills --> Set
10. Certifications --> List
11. Address --> dictionary
        street --> string
        city --> string
        state --> string
        zipcode --> string

```python
candidate = {
    "first_name" : "Deepak",
    "last_name" : "Veeravalli",
    "email" : "test@test.com",
    "phone" : "+1-123456789",
    "gender" : ('male'),
    "linkedin_url" : "https://linkedin/dveeravalli",
    "qualifications" : {'Msc,IS','BCA'},
    "skills" : {
        "database" : {"oracle","DB2","Sybase"},
        "programming" : {"c","c++","java","e-script","apex","python","HTML","JavaScript","css"},
        "management" : {"teamlead","project management"},
        "ai" : {"Agentic AI","langchain","langgrpah","n8n","crew.ai"}
    },
    "certifications": ["salesforce","siebel"],
    "address" : {
        "street" : "123 random Street",
        "city" : "random city",
        "state" : "random state",
        "zip" : "CA 12345"
    }
}

print(candidate['address'])
print(candidate['address']['zip'])
print(candidate.keys())
print(len(candidate.keys()))
print(candidate.values())
```

# accessing dictionary using for loop

1. Accessing Keys only
2. Accessing Values only
3. Accessing both keys and values

```python
for key in candidate:
    print(key)
```

```python
for value in candidate.values():
    print(value)
```

```python
for key,value in candidate.items():
    print(f"The key is: {key} and the value is: {value}")
```
